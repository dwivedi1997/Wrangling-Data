**Using datetime framework of Pandas**

```
ipl['Match_No']=[i for i in range(1,ipl.shape[0]+1)]
ipl['Date']=pd.to_datetime(ipl.Date)
ipl['Year']=ipl['Date'].dt.year
ipl['Month_No']=ipl['Date'].dt.month
```

Converting month number into month

```
month={1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
month_name=[]
for j in ipl.Month_No:
  month_name.append(month[j])
ipl['Month']=month_name
```

Pandas Datetime

Dividing Teams into two parts

```
team1=[]
team2=[]
for i in ipl.Match:
  team1.append(i[i.find(':')+2:i.find('v')-1])
  team2.append(i[i.find('v')+2:])
ipl['Team_1']=team1
ipl['Team_2']=team2
```

Classifying data into Mode by which match has been won

```
import re
mode=[]
for i in ipl.Result:
  if len(i.split())>2:
    mode.append(i.split()[-1])
  else:
    mode.append(i)
ipl['Winning_Mode']=mode
```

Choosing Winning Team

```
winning=[]
for i in ipl.Result:
  if i=='eliminator)':
    winning.append(i[i.find('(')+1,:i.find('w')-1])
  else:
    winning.append(i[:i.find('w')-1])
ipl['Winning_Team']=winning
```

Selecting the Margin by which Match has been Won

```
margin=[]
for i in ipl.Result:
  if i=='No Result':
    margin.append('No Result')
  elif i.split()[-1]=='eliminator)':
    margin.append('Won by Eliminator')
  else:
    margin.append(i.split()[-2])
ipl['Winning_Margin']=margin
```

Selecting Which Team has Bat first

```
first=[]
for i in range(len(ipl.Winning_Mode)):
  if ipl.Winning_Mode[i]=='Wickets':
    if ipl.Winning_Team[i]==ipl.Team_1[i]:
      first.append(ipl.Team_2[i])
    else:
      first.append(ipl.Team_1[i])
    
  else:
    first.append(ipl.Winning_Team[i])  
ipl['Team_Batting_First']=first
```
